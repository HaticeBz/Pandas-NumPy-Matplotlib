```python
#Exercise - 2
import pandas as pd

# I created a sample DataFrame
data = {
    'gender': [1, 2, 1, 2, 1],
    'situation': [1, 2, 3, 4, 1],
    'tea': [2.5, 1.0, 3.0, 0.5, 2.0],
    'coffee': [1.0, 3.0, 2.0, 4.0, 1.5],
    'height': [175.0, 160.0, 180.0, 165.0, 170.0],
    'weight': [70, 60, 80, 65, 75],
    'age': [30, 25, 40, 35, 50],
    'meat': [5, 3, 2, 4, 1],
    'fish': [4, 2, 5, 3, 1],
    'raw_fruit': [3, 4, 2, 1, 5],
    'cooked_fruit_veg': [5, 1, 2, 3, 4],
    'chocol': [1, 2, 3, 4, 5],
    'fat': [5, 6, 7, 4, 3]
}

df = pd.DataFrame(data)

# I renamed the columns
df = df.rename(columns={
    'gender': 'sex',
    'situation': 'marital_status',
    'tea': 'daily_tea_consumption',
    'coffee': 'daily_coffee_consumption',
    'height': 'body_height',
    'weight': 'body_weight',
    'age': 'current_age',
    'meat': 'meat_consumption',
    'fish': 'fish_consumption',
    'raw_fruit': 'raw_fruit_consumption',
    'cooked_fruit_veg': 'cooked_fruit_veg_consumption',
    'chocol': 'chocolate_consumption',
    'fat': 'fat_consumption'
})

# I changed data types
df['sex'] = df['sex'].astype('category')
df['marital_status'] = df['marital_status'].astype('category')
df['daily_tea_consumption'] = df['daily_tea_consumption'].astype('float')
df['daily_coffee_consumption'] = df['daily_coffee_consumption'].astype('float')
df['body_height'] = df['body_height'].astype('float')
df['body_weight'] = df['body_weight'].astype('int')
df['current_age'] = df['current_age'].astype('int')
df['meat_consumption'] = df['meat_consumption'].astype('category')
df['fish_consumption'] = df['fish_consumption'].astype('category')
df['raw_fruit_consumption'] = df['raw_fruit_consumption'].astype('category')
df['cooked_fruit_veg_consumption'] = df['cooked_fruit_veg_consumption'].astype('category')
df['chocolate_consumption'] = df['chocolate_consumption'].astype('category')
df['fat_consumption'] = df['fat_consumption'].astype('category')

print(df)

# I created a CSV file
df.to_csv('modified_data.csv', index=False)
print("csv file created successfully!")
```

```python
#Exercise - 3
student_scores = {           #melt method
    'Student': [1, 2, 3, 4, 5],
    'Before' : [75, 30, 100, 50, 60],
    'After' : [85, 50, 100, 52, 65]
}

df = pd.DataFrame(student_scores)

df_melted = pd.melt(df, id_vars = 'Student', value_vars = ['Before','After'], var_name = 'Time', value_name = 'Score')

print(df_melted)
```

```python
#Exercise - 3
student_scores = {        #stack method
    'Student': [1, 2, 3, 4, 5],
    'Before' : [75, 30, 100, 50, 60],
    'After' : [85, 50, 100, 52, 65]
}

df = pd.DataFrame(student_scores)

df_stacked = df.stack()
print(df_stacked)
```

```python
#Exercise - 3
student_scores = {         #unstack method: The stack method should be applied first. In the end, it returns to its original state.
    'Student': [1, 2, 3, 4, 5],
    'Before' : [75, 30, 100, 50, 60],
    'After' : [85, 50, 100, 52, 65]
}

df = pd.DataFrame(student_scores)

df_stacked = df.stack()

df_unstacked = df_stacked.unstack()
print(df_unstacked)
```

```python
#Exercise - 4
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Situation': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
})

proportions = df.groupby(['Situation', 'Gender']).size().unstack()
proportions = proportions.div(proportions.sum(axis=1), axis=0)

fig, ax = plt.subplots(1, 1, figsize=(8, 4))

proportions.plot(kind='bar', stacked=True, ax=ax, color=['blue', 'pink'])

ax.set_title('Proportion of Males and Females in Each Situation')
ax.set_ylabel('Proportion')
ax.set_xlabel('Situation')
plt.tight_layout()
plt.show()
```

```python
#Exercise 5 - a and b
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Class']
data = pd.read_csv(url, names=names)

# Here I created a boxplot of 'Petal.Length' for each species
plt.figure(figsize=(8, 4))
plt.boxplot([data[data['Class'] == 'Iris-setosa']['Petal.Length'],
             data[data['Class'] == 'Iris-versicolor']['Petal.Length'],
             data[data['Class'] == 'Iris-virginica']['Petal.Length']],
            labels=['Setosa', 'Versicolor', 'Virginica'])
plt.title('Boxplot of Petal.Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal.Length')
plt.show()
```

```python
#Exercise 5 - d
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Class']
data = pd.read_csv(url, names=names)

species_colors = {'Iris-setosa': 'blue', 'Iris-versicolor': 'green', 'Iris-virginica': 'red'}  # I gave colors to the species

plt.figure(figsize=(8, 4))  # I created a scatterplot
for species, group in data.groupby('Class'):
    plt.scatter(group['Sepal.Length'], group['Petal.Length'], label=species, color=species_colors[species])
plt.title('Scatterplot of Sepal.Length vs Petal.Length')
plt.xlabel('Petal.Length')
plt.ylabel('Sepal.Length')
plt.legend(title='Species')
plt.show()
```

```python
#Exercise 5 - e
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Class']
data = pd.read_csv(url, names=names)

# I created a kernel density plot
plt.figure(figsize=(8, 4))
sns.kdeplot(data[data['Class'] == 'Iris-setosa']['Petal.Length'], label='Setosa', color='blue', fill=True)  #I colored under the curves with fill=True
sns.kdeplot(data[data['Class'] == 'Iris-versicolor']['Petal.Length'], label='Versicolor', color='green', fill=True)
sns.kdeplot(data[data['Class'] == 'Iris-virginica']['Petal.Length'], label='Virginica', color='red', fill=True)
plt.title('Kernel Density Plot of Petal.Length by Species')
plt.xlabel('Petal.Length')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#Exercise 6 - b
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt

data = sm.datasets.get_rdataset('EuStockMarkets', 'datasets').data

print(data.head())

#Let's add a date range (let's say the start date is January 1991)
dates = pd.date_range(start='1991-01-01', periods=len(data), freq='B')  #B = workday frequency

#Here I added the date to the index
data.index = dates
print(data.head())

#dictionary that maps column names to colors
color_dict = {'DAX': 'blue', 'SMI': 'green', 'CAC': 'red', 'FTSE': 'purple'}
plt.figure(figsize=(8, 4))

#plot each column according to the color in color_dict
for column in data.columns:
  plt.plot(data.index, data[column], color=color_dict[column], label=column)

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Stock Indices for Various European Stock Markets')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
